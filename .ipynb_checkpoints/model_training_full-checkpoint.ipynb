```python
from keras.models import Sequential
from keras.models import load_model
from keras.layers.core import Dense
from keras.layers.core import Dropout
from keras.callbacks import EarlyStopping
from keras import optimizers
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from tensorflow import keras
from pandas import DataFrame as df

import pandas as pd
import numpy
import tensorflow as tf
```

```python
work_dir='/home/jwseo/drug_repositioning/new_workflow/'
```

```python
#Loading Interaction data
interaction_loc = work_dir + "final_interaction.txt"
Interaction_original = pd.read_csv(interaction_loc)
Interaction_original
```

```python
ssp = pd.read_csv(work_dir+"SSP_pca.txt", delimiter = "\t")
ssp
```

```python
psp = pd.read_csv(work_dir"PSP_pca.txt", delimiter = "\t")
psp
```

```python
#ssp와 psp를 dictionary 형태로 바꿈(빨리 찾기 위해서)
ssp_dic = {}
for i in range(0, len(ssp)) :
    drug = ssp.iloc[i,0]
    ssp_temp = list(ssp.iloc[i,1:202])
    ssp_dic[drug] = ssp_temp
    
psp_dic = {} 
for i in range(0, len(psp)) :
    drug = psp.iloc[i,0]
    psp_temp = list(psp.iloc[i,1:302])
    psp_dic[drug] = psp_temp
```

```python
#drug pair의 ssp와 psp 가져와서 input_data 만듬
ddi_input = []
for i in range(0, len(Interaction_original)) :
    drug_a = Interaction_original.iloc[i,0]
    drug_b = Interaction_original.iloc[i,1]
    ddi_type = Interaction_original.iloc[i,2]
    if drug_a in ssp_dic.keys() and drug_a in psp_dic.keys() :
        if drug_b in ssp_dic.keys() and drug_b in psp_dic.keys() : 
            drug_a_ssp = ssp_dic[drug_a]
            drug_a_psp = psp_dic[drug_a]
            drug_b_ssp = ssp_dic[drug_b]
            drug_b_psp = psp_dic[drug_b]
            input_temp = [drug_a]+[drug_b]+drug_a_ssp + drug_a_psp + drug_b_ssp + drug_b_psp + [ddi_type]
            ddi_input.append(input_temp
dataframe = df(ddi_input)
```

```python
#Input 저장할거면 돌리기
#dataframe.to_csv(work_dir+"model_input.csv",index=None)
```

```python
#drug pair, 모델의 X,Y로 나눠줌
dataset = dataframe.values
drug_pair = dataset[:,0:2]
X = dataset[:,2:1002].astype(float)
Y_obj = dataset[:,1002]
```

```python
#Y값에 대해서 multilabel 형태로 바꿔줌
e = LabelEncoder()
e.fit(Y_obj)
Y = e.transform(Y_obj)
Y_encoded = tf.keras.utils.to_categorical(Y)
```

```python
#train test split
drug_train, drug_test, X_train, X_test, Y_train, Y_test = train_test_split(drug_pair, X, Y_encoded, test_size=0.25, random_state=10)
```

```python
model = Sequential()
model.add(Dense(1024, input_dim=1000, activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(1024, activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(1024, activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(80, activation='softmax'))

adam = tf.optimizers.Adam(0.01)
model.compile(loss='categorical_crossentropy',
            optimizer='adam',
            metrics=['accuracy'])
```

```python
#model 구조
model.summary()
```

```python
model.fit(X_train, Y_train, epochs=100, batch_size=128)
print("Test Set Accuracy : %.4f"%(model.evaluate(X_test, Y_test)[1]))
```

```python
#Saving Model
model_name = "model_parameter.h5" 
model.save(work_dir+model_name)
```
